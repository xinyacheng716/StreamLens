```python
import pandas as pd

movies = pd.read_csv('ml-latest-small/movies.csv')
ratings = pd.read_csv('ml-latest-small/ratings.csv')

# shape 回傳（行數，欄數）
print("movies shape:", movies.shape) #(9742部電影, 3欄位)
print("ratings shape:", ratings.shape) #(100,836部電影, 4欄位)
```

```python
print("=== movies.csv ===")
print(movies.head()) #欄位樣式default前五列

print("\n=== ratings.csv ===")
print(ratings.head())
```

```python
print("=== movies info ===")
print(movies.info()) #每欄位資料型態以及缺失值 #title and genre are str not list yet

print("\n=== ratings info ===")
print(ratings.info())
```

```python
# Check rating distribution
print("=== Rating Distribution ===")
print(ratings['rating'].value_counts().sort_index())

print("\nUnique users:", ratings['userId'].nunique())
print("Unique movies rated:", ratings['movieId'].nunique())

# 評分結果為右偏分配，代表大部分人傾向給高分，所以光看平均分不夠，因為整體分數本就偏高
# unique movie = 9724 但是movieId電影有 9742 部代表有18部電影沒有人平分過
# users目前較少只有六百多位但後續換成完整資料即可解決
```

```python
# Check how many unique genres exist
# First, split the pipe-separated genres and collect all unique values

all_genres = movies['genres'].str.split('|').explode().unique()
# split('|')用來把str的字串切開變成list ; 再把 list 展開成個別的行 ; 最後取不重複值用.unique()
print("All unique genres:")
print(sorted(all_genres))
print("\nTotal unique genres:", len(all_genres))

# no genres listed isn't a genre, make sure to delete it when analyze; also iMAX is not a genre as well
```
